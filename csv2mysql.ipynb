## Import Libraries

```python
import pandas as pd
import pymysql
```

## Data Loading and Preprocessing

```python
crime_data = pd.DataFrame(columns=[
    "Latitude", "Longitude", "Crime_Description", "Sub_Category",
    "City", "CrimeDate", "DateYear", "DateMonth"
])
```

```python
# Chicago

chicago_data = pd.read_csv("ChicagoCrimes.csv")

chicago_data['CrimeDate'] = chicago_data['Date'].str.split().str[0]
chicago_data["CrimeDate"] = pd.to_datetime(chicago_data["CrimeDate"])

mapped_chicago_data = chicago_data.assign(
    Crime_Description=chicago_data["Description"],  
    Sub_Category= chicago_data["PrimaryType"],  
    City="Chicago", 
    CrimeDate=chicago_data["CrimeDate"], 
    DateYear=chicago_data["CrimeDate"].dt.year,  
    DateMonth=chicago_data["CrimeDate"].dt.month,  
)[["Latitude", "Longitude", "Crime_Description", "Sub_Category", "City", "CrimeDate", "DateYear", "DateMonth"]]

crime_data = pd.concat([crime_data, mapped_chicago_data], ignore_index=True)

crime_data
```

```python
# San Francisco

sf_data = pd.read_csv("SanFranciscoCrimes.csv")

sf_data["Incident Date"] = pd.to_datetime(sf_data["Incident Date"])

mapped_sf_data = sf_data.assign(
    Crime_Description=sf_data["Incident Description"],  
    Sub_Category= sf_data["Incident Category"],  
    City="San Francisco",  
    CrimeDate=sf_data["Incident Date"],  
    DateYear=sf_data["Incident Year"],  
    DateMonth=sf_data["Incident Date"].dt.month, 
)[["Latitude", "Longitude", "Crime_Description", "Sub_Category", "City", "CrimeDate", "DateYear", "DateMonth"]]

crime_data = pd.concat([crime_data, mapped_sf_data], ignore_index=True)

crime_data
```

```python
# Seattle

seattle_data = pd.read_csv("SeattleCrimes.csv")

seattle_data["CrimeDate"] = seattle_data["Report DateTime"].str.split().str[0]
seattle_data["CrimeDate"] = pd.to_datetime(seattle_data["CrimeDate"])

mapped_seattle_data = seattle_data.assign(
    Crime_Description=seattle_data["Offense"], 
    Sub_Category= seattle_data["Offense Parent Group"],  
    City="Seattle", 
    CrimeDate=seattle_data["CrimeDate"],  
    DateYear=seattle_data["CrimeDate"].dt.year, 
    DateMonth=seattle_data["CrimeDate"].dt.month, 
)[["Latitude", "Longitude", "Crime_Description", "Sub_Category", "City", "CrimeDate", "DateYear", "DateMonth"]]

crime_data = pd.concat([crime_data, mapped_seattle_data], ignore_index=True)

crime_data
```

```python
#drop rows with any missing values in crime_data
crime_data.dropna(axis=0, inplace=True) 

#filter rows from 2019 onwards
crime_data = crime_data[crime_data['CrimeDate'] >= '2019-01-01']
```

```python
# Crime Categories Classification
def categorize_crime(crime_category):
    crime_category = str(crime_category).lower()

    if any(keyword in crime_category for keyword in [
        'felony', 'murder', 'manslaughter', 'homicide', 'missing', 'assault',
        'battery', 'kidnapping', 'abduction', 'child', 'children'
    ]):
        return 'Violent Crimes'

    elif any(keyword in crime_category for keyword in [
        'sex', 'sexual', 'prostitue', 'prostitution', 'rape', 
        'pornography', 'family', 'domestic'
    ]):
        return 'Domestic and Sexual Offenses'

    elif any(keyword in crime_category for keyword in [
        'vandalism', 'trespass', 'trespassing',
        'retail', 'property', 'larceny'
    ]):
        return 'Property Crimes'
    
    elif any(keyword in crime_category for keyword in [
        'burglary'
    ]):
        return 'Burglary'

    elif any(keyword in crime_category for keyword in [
       'narcotics', 'drugs', 'heroin', 
       'marjuana', 'drug', 'cannabis', 'liquor'
    ]):
        return 'Drug Offenses'

    elif any(keyword in crime_category for keyword in [
        'fraud', 'card', 'forgery', 'pickpocket', 'checks',
        'robbery', 'theft', 'purse', 'financial', "counterfeit",
        'bribery', 'blackmail', 'embezzlement'
    ]):
        return 'Theft and Fraud'

    elif any(keyword in crime_category for keyword in [
        'weapons', 'armed', 'carrying', ' ammunition', 
        'gun', 'weapon', 'firearms', 'possession', 'knife'
    ]):
        return 'Weapons Offenses'

    elif any(keyword in crime_category for keyword in [
         'speed', 'traffic', 'bike', 'driving', 'car', 'vehicle',
         'automobile', 'drunk', 'bus'
    ]):
        return 'Traffic and Vehicular Crimes'

    elif any(keyword in crime_category for keyword in [
        'obscenity', 'restraining', 'predatory', 'stalking', 'gambling', 'warrant',
        'public', 'entry', 'harrassment', 'restraint', 'peeping', 'damage', 'intimidation'
    ]):
        return 'Public Order Offenses'

    elif any(keyword in crime_category for keyword in [
        'obstruct', 'arson', 'bomb', 'school', 'riot', 'case',
        'suspicious', 'violation', 'fire', 'civil', 'cruelty',
        'law', 'conspiracy', 'illegal', 'state', 'police'
    ]):
        return 'Crimes Against Government'

    else:
        return 'Other'

crime_data['Crime_Category'] = crime_data['Sub_Category'].apply(categorize_crime)

```

```python
crime_data['CrimeDate'] = crime_data['CrimeDate'].dt.strftime('%Y-%m-%d %H:%M:%S')
```

```python
def create_crimes_table(cursor):
    cursor.execute("DROP TABLE IF EXISTS hovetl_crimes")
    create_table_query = """
        CREATE TABLE IF NOT EXISTS hovetl_crimes (
            CID INT PRIMARY KEY AUTO_INCREMENT,
            CrimeDate DATETIME,
            DateYear INT,
            DateMonth INT,
            City VARCHAR(100),
            Crime_Category VARCHAR(100),
            Sub_Category VARCHAR(100),
            Crime_Description VARCHAR(100),
            Latitude FLOAT,
            Longitude FLOAT
        ) ENGINE=MyISAM DEFAULT CHARSET=latin1 AUTO_INCREMENT=1 ;
    """
    cursor.execute(create_table_query)
    print("Crimes table created.")
```

```python
def insert_data_to_mysql(cursor, dataframe, table_name="hovetl_crimes"):

    insert_query = f"""
        INSERT INTO {table_name} ( 	
            Latitude, Longitude, Crime_Description,	Sub_Category,	
            City, CrimeDate, DateYear, DateMonth, Crime_Category
        ) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s)
    """

    data_tuples = dataframe.to_records(index=False).tolist()
    
    try:
        cursor.executemany(insert_query, data_tuples)
        print(f"Successfully inserted {len(data_tuples)} records into {table_name}.")
    except Exception as e:
        print(f"Error inserting data: {e}")
```

```python
def main():
    create_crimes_table(cur)

    insert_data_to_mysql(cur, crime_data)

    cur.close()
    conn.close()


if __name__ == "__main__":
    main()

```
